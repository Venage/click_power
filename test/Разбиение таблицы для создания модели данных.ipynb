```python
import pandas as pd


df =  pd.read_csv(r'C:\Users\Venage\PycharmProjects\click_power\test\dashminox_razbit.csv')  # Или другой формат
# ['Тип доставки', 'Номер', 'Менеджер', 'Статус заказа', 'Город',
#        'Тип оплаты', 'Промокод', 'Наложенный платёж', 'date', 'hour', 'day',
#        'day_of_week', 'Количество_полное', 'Выручка', 'sku_clear', 'sku_nabor',
#        'Себестоимость', 'Маржа', 'Ссылка', 'Бренд', 'СТМ',
#        'Комментарий Кладовщика 2', 'Контактный телефон', 'E-mail', 'Отчество',
#        'Имя', 'Фамилия', 'Индекс', 'Оплачено', 'Склад', 'Дата отгрузки',
#        'Причина отмены заказа', 'Канал', 'Источник', 'Кампания',
#        'Содержание кампании', 'Первый заказ', 'Последний заказ',
#        'Ключевое слово', 'Порядковый номер заказа у клиента',
#        'Итоговая скидка', 'Передача данных в службу доставки',
#        'Себестоимость доставки', 'Адрес доставки', 'Стоимость доставки',
#        'Адрес ПВЗ', 'Регион', 'Способ оформления', 'Группа статусов',
#        'Внутренний ID', 'UTM метка', 'YaMetraki ClientID',
#        'Работа с базой - отметка', 'Помощь при оформлении', 'Работа проведена',
#        'Откуда продажа', 'Название']

df = df.dropna(subset=['Номер'])
# --- Создание таблицы "Клиенты" ---
df_клиенты = df[['Контактный телефон', 'E-mail', 'Отчество', 'Имя', 'Фамилия', 'Индекс', 'Город', 'Регион', 'Адрес доставки', 'Адрес ПВЗ', 'YaMetraki ClientID']].drop_duplicates(subset=['Контактный телефон', 'E-mail', 'Отчество', 'Имя', 'Фамилия'], keep='first').copy()
df_клиенты['ClientId'] = range(1, len(df_клиенты) + 1)  # Создаем уникальный ClientId



# --- Создание таблицы "Заказы" ---
df_заказы = df[['Номер', 'date', 'hour', 'day', 'day_of_week', 'Количество_полное', 'Выручка', 'Себестоимость', 'Маржа', 'Оплачено', 'Итоговая скидка', 'Себестоимость доставки', 'Стоимость доставки', 'Наложенный платёж', 'Тип оплаты', 'Промокод', 'Тип доставки', 'Внутренний ID', 'Передача данных в службу доставки', 'Статус заказа', 'Склад', 'Первый заказ', 'Последний заказ', 'Порядковый номер заказа у клиента', 'Способ оформления', 'Контактный телефон','Менеджер','Причина отмены заказа','Склад','Статус заказа','Тип доставки','Тип оплаты','Промокод','Канал']].copy()
# Добавляем ClientId на основе соответствия 'Контактный телефон' в обеих таблицах
df_заказы = pd.merge(df_заказы, df_клиенты[['Контактный телефон', 'ClientId']], on='Контактный телефон', how='left')

# --- Создание таблицы "Товары" ---
df_товары = df[['sku_clear', 'sku_nabor', 'Бренд', 'СТМ', 'Ссылка', 'Название']].drop_duplicates(subset=['sku_clear']).copy()

# --- Создание таблицы "Менеджеры" ---
df_менеджеры = df[['Менеджер']].drop_duplicates(subset=['Менеджер']).copy()
df_менеджеры['МенеджерId'] = range(1, len(df_менеджеры) + 1)

# --- Создание таблицы "Каналы" ---
df_каналы = df[['Канал']].drop_duplicates(subset=['Канал']).copy()
df_каналы['КаналId'] = range(1, len(df_каналы) + 1)


# --- Создание таблицы "Типы оплаты" ---
df_типы_оплаты = df[['Тип оплаты']].drop_duplicates(subset=['Тип оплаты']).copy()
df_типы_оплаты['ТипОплатыId'] = range(1, len(df_типы_оплаты) + 1)

# --- Создание таблицы "Промокоды" ---
df_промокоды = df[['Промокод']].drop_duplicates(subset=['Промокод']).copy()
df_промокоды['ПромокодId'] = range(1, len(df_промокоды) + 1)

# --- Создание таблицы "Типы доставки" ---
df_типы_доставки = df[['Тип доставки']].drop_duplicates(subset=['Тип доставки']).copy()
df_типы_доставки['ТипДоставкиId'] = range(1, len(df_типы_доставки) + 1)

# --- Создание таблицы "Статусы заказов" ---
df_статусы_заказов = df[['Статус заказа', 'Группа статусов']].drop_duplicates(subset=['Статус заказа', 'Группа статусов']).copy()
df_статусы_заказов['СтатусЗаказаId'] = range(1, len(df_статусы_заказов) + 1)

# --- Создание таблицы "Склады" ---
df_склады = df[['Склад']].drop_duplicates(subset=['Склад']).copy()
df_склады['СкладId'] = range(1, len(df_склады) + 1)

# --- Создание таблицы "Причины отмены заказа" ---
df_причины_отмены = df[['Причина отмены заказа']].drop_duplicates(subset=['Причина отмены заказа']).copy()
df_причины_отмены['ПричинаОтменыId'] = range(1, len(df_причины_отмены) + 1)

# --- Подготовка таблицы заказов с внешними ключами ---
# Merging с таблицами чтобы добавить ID
df_заказы = pd.merge(df_заказы, df_менеджеры[['Менеджер', 'МенеджерId']], on='Менеджер', how='left')
df_заказы = pd.merge(df_заказы, df_каналы[['Канал', 'КаналId']], on='Канал', how='left')
# df_заказы = pd.merge(df_заказы, df_типы_оплаты[['Тип оплаты', 'ТипОплатыId']], on='Тип оплаты', how='left')
df_заказы = pd.merge(df_заказы, df_промокоды[['Промокод', 'ПромокодId']], on='Промокод', how='left')
df_заказы = pd.merge(df_заказы, df_типы_доставки[['Тип доставки', 'ТипДоставкиId']], on='Тип доставки', how='left')
df_заказы = pd.merge(df_заказы, df_статусы_заказов[['Статус заказа', 'СтатусЗаказаId']], on='Статус заказа', how='left')
df_заказы = pd.merge(df_заказы, df_склады[['Склад', 'СкладId']], on='Склад', how='left')
# df_заказы = pd.merge(df_заказы, df_причины_отмены[['Причина отмены заказа', 'ПричинаОтменыId']], on='Причина отмены заказа', how='left')
df_заказы

# # --- Сохраните таблицы в отдельные файлы CSV или другом формате ---
# df_клиенты.to_csv("клиенты.csv", index=False)
# df_заказы.to_csv("заказы.csv", index=False)
# df_товары.to_csv("товары.csv", index=False)
# df_менеджеры.to_csv("менеджеры.csv", index=False)
# df_каналы.to_csv("каналы.csv", index=False)
# df_типы_оплаты.to_csv("типы_оплаты.csv", index=False)
# df_промокоды.to_csv("промокоды.csv", index=False)
# df_типы_доставки.to_csv("типы_доставки.csv", index=False)
# df_статусы_заказов.to_csv("статусы_заказов.csv", index=False)
# df_склады.to_csv("склады.csv", index=False)
# df_причины_отмены.to_csv("причины_отмены.csv", index=False)
# 
# print("Таблицы успешно разделены и сохранены!")
```
